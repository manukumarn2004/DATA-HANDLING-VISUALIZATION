```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
print("step 1")
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,50000,100)
print(x)
ax.plot(x,np.sin(x))
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,500,100)
print(x)
ax.plot(x,np.cos(x))
ax.plot(x,np.sin(x))
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,500,100)
print(x)
ax.plot(x,np.cos(x))
ax.plot(x,np.sin(x))
ax.plot(x,np.tan(x))
```

```python

ax=fig=plt.figure()
plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.sin(x))
ax.set_title('Simple plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.sin(x),label='sin')
ax.plot(x,np.cos(x),label='cos')
ax.set_title('multiple lines')
ax.set_xlabel('periods')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.sin(x),label='sin',color='green')
ax.plot(x,np.cos(x),label='cos',color='red')
ax.set_title('multiple lines')
ax.set_xlabel('periods')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.sin(x),label='sin',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',linestyle='dotted')
ax.plot(x,np.sin(x+1),label='sin x+1',linestyle='dashdot')
ax.set_title('multiple lines')
ax.set_xlabel('periods')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```
